## Results visualization

```r
bonferroni <- log10(0.05/row(resultGEMMA))
```

```r
png("GWASresults_manhattan.png")
manhattan(resultGEMMA, chr="chr", bp="ps", p="p_lrt", snp="rs", genomewideline = bonferroni)
dev.off()
```

```r
## QQ plot
png("GWAS_casecontrol_opiod_QQplot.png")
    qq(resultGEMMA$p_lrt)
dev.off()
```

## Extract top results

```r
resultGEMMA %>%
    mutate(negLogP = -log10(p_lrt)) %>%
    select(chr, rs, p_lrt, negLogP) %>%
    filter(negLogP > 5)
```
